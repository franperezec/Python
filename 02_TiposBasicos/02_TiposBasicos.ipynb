## Cadenas de Texto (Strings) en Python

### 쯈u칠 es una Cadena de Texto?

Una cadena de texto, o *string*, es una secuencia de caracteres encerrada entre comillas. En Python, las cadenas de texto pueden ser delimitadas por comillas simples (`'...'`), comillas dobles (`"..."`), o comillas triples (`'''...'''` o `"""..."""`). Las comillas triples permiten incluir saltos de l칤nea dentro de la cadena.

### Ejemplos de Cadenas de Texto

```python
nombre_persona = "Francisco"
direccion = """Calle 123,
Colonia Centro, Quito"""
```

En este ejemplo, `nombre_persona` es una cadena simple y `direccion` es una cadena que ocupa varias l칤neas, gracias a las comillas triples.

### Operaciones con Cadenas de Texto

1. **Concatenaci칩n**: Puedes unir cadenas de texto usando el operador `+` o simplemente imprimiendo varias cadenas juntas.
   
   ```python
   print(nombre_persona, direccion)
   ```

2. **Longitud de la Cadena**: La funci칩n `len()` devuelve el n칰mero de caracteres en una cadena.

   ```python
   print(len(nombre_persona))  # Resultado: 9
   ```

3. **Acceso a Caracteres**: Puedes acceder a un car치cter espec칤fico de una cadena usando 칤ndices. Python utiliza 칤ndices basados en cero, lo que significa que el primer car치cter tiene el 칤ndice 0.

   ```python
   print(nombre_persona[0])  # Resultado: "F"
   ```

4. **Subcadenas**: Puedes extraer una parte de la cadena utilizando el operador de corte (slice), especificando un rango de 칤ndices.

   ```python
   print(direccion[0:7])  # Resultado: "Calle 1"
   print(direccion[9:])   # Resultado: "Colonia Centro, Quito"
   print(direccion[:7])   # Resultado: "Calle 1"
   print(direccion[-5:])  # Resultado: "Quito"
   print(direccion[:-5])  # Resultado: "Calle 123,\nColonia Centro, "
   ```

5. **Corte con Pasos**: Puedes tambi칠n especificar un tercer par치metro en el corte para definir el tama침o del paso, es decir, cu치ntos caracteres saltar cada vez.

   ```python
   print(direccion[::2])  # Resultado: "Cae13\nCooaCno ut"
   ```

6. **Imprimir la Cadena Completa**: Usar el operador de corte sin especificar el inicio o el final imprimir치 toda la cadena.

   ```python
   print(direccion[:])  # Resultado: "Calle 123,\nColonia Centro, Quito"
   ```

### Resumen de Operaciones con Strings

- **Acceso por 칤ndice**: `[칤ndice]`
- **Subcadena**: `[inicio:fin]`
- **Subcadena con paso**: `[inicio:fin:paso]`
- **Longitud**: `len(cadena)`

Estas herramientas te permiten manipular y trabajar con cadenas de texto de manera efectiva en Python. Aprender a utilizar estas t칠cnicas es fundamental para manejar texto en tus programas.

## Formato de Strings en Python

Las cadenas de texto (strings) en Python pueden formatearse y manipularse de diversas maneras. Una de las operaciones m치s comunes es la **concatenaci칩n**, que consiste en unir varias cadenas para formar una nueva.

### Concatenaci칩n de Strings

La forma m치s sencilla de concatenar (unir) strings es usando el operador `+`.

```python
nombre = "Francisco"
apellido = "P칠rez"
nombre_completo = nombre + " " + apellido  # Concatenaci칩n de strings
print(nombre_completo)  # Resultado: "Francisco P칠rez"
```

En este ejemplo, las variables `nombre` y `apellido` se concatenan junto con un espacio (`" "`) para formar el string `nombre_completo`.

### Uso de f-Strings para Formatear Strings

Python ofrece una forma m치s potente y flexible de formatear strings mediante **f-strings** (cadenas formateadas). Las f-strings permiten incrustar expresiones dentro de una cadena utilizando llaves `{}`. Para usar una f-string, simplemente coloca una `f` antes de las comillas.

```python
nombre_completo = f"{nombre} {apellido}"
print(nombre_completo)  # Resultado: "Francisco P칠rez"
```

En este ejemplo, el c칩digo dentro de las llaves `{}` se eval칰a y se inserta en la cadena resultante.

### Ejemplo Avanzado con f-Strings

Las f-strings no solo permiten la interpolaci칩n de variables, sino que tambi칠n puedes incluir expresiones y operaciones directamente dentro de las llaves.

```python
# Mostrar el primer car치cter del nombre completo y un c치lculo
print(f"{nombre_completo[0]} es el primer caracter de {nombre_completo} {1+2}")
```

Este c칩digo produce la salida:

```
F es el primer caracter de Francisco P칠rez 3
```

Aqu칤:
- `nombre_completo[0]` accede al primer car치cter de la cadena `nombre_completo`, que es `"F"`.
- Dentro de las llaves `{1+2}`, se realiza la suma `1 + 2`, y el resultado `3` se inserta en la cadena.

### Ventajas de Usar f-Strings

- **Legibilidad**: El c칩digo es m치s f치cil de leer y entender.
- **Eficiencia**: Las f-strings son m치s r치pidas que las operaciones tradicionales de concatenaci칩n.
- **Flexibilidad**: Puedes insertar expresiones complejas directamente dentro de las cadenas.

Las f-strings son una herramienta poderosa en Python para formatear y manipular texto de manera eficiente y clara.

## M칠todos de String en Python

En Python, los strings tienen varios m칠todos que permiten manipular y transformar el texto de diversas maneras. A continuaci칩n, se muestran algunos de los m칠todos m치s comunes y c칩mo utilizarlos.

```python
animal = "gato tricolor"

# Transforma todo el string a may칰sculas
print(animal.upper())  # GATO TRICOLOR

# Transforma todo el string a min칰sculas
print(animal.lower())  # gato tricolor

# Convierte la primera letra del string a may칰sculas
print(animal.capitalize())  # Gato tricolor

# Convierte la primera letra de cada palabra a may칰sculas
print(animal.title())  # Gato Tricolor

# Invierte las may칰sculas y min칰sculas en el string
print(animal.swapcase())  # GATO TRICOLOR

# Reemplaza la palabra "gato" por "perro"
print(animal.replace("gato", "perro"))  # perro tricolor

# Elimina espacios en blanco al principio y al final del string
print(animal.strip())  # gato tricolor

# Elimina espacios y luego convierte la primera letra a may칰sculas
print(animal.strip().capitalize())  # Gato tricolor

# Elimina espacios en blanco a la izquierda
print(animal.lstrip())  # gato tricolor

# Elimina espacios en blanco a la derecha
print(animal.rstrip())  # gato tricolor

# Devuelve la posici칩n de la primera aparici칩n de "a"
print(animal.find("a"))  # 1

# Devuelve True si "gato" est치 en el string
print("gato" in animal)  # True

# Devuelve True si "gato" no est치 en el string
print("gato" not in animal)  # False

# Reemplaza la palabra "gato" por "perro"
print(animal.replace("gato", "perro"))  # perro tricolor

# Devuelve la longitud del string
print(len(animal))  # 13

# Divide el string en una lista de palabras
print(animal.split())  # ['gato', 'tricolor']

# Cuenta cu치ntas veces aparece "a" en el string
print(animal.count("a"))  # 2

# Devuelve la posici칩n de la 칰ltima aparici칩n de "a"
print(animal.rfind("a"))  # 10

# Devuelve la posici칩n de la primera aparici칩n de "a"
print(animal.index("a"))  # 1

# Devuelve la posici칩n de la 칰ltima aparici칩n de "a"
print(animal.rindex("a"))  # 10

# Devuelve True si el string empieza con "gato"
print(animal.startswith("gato"))  # True

# Devuelve True si el string termina con "tricolor"
print(animal.endswith("tricolor"))  # True

# Devuelve True si el string es alfanum칠rico (letras y n칰meros)
print(animal.isalnum())  # False

# Devuelve True si el string es alfab칠tico (solo letras)
print(animal.isalpha())  # False

# Devuelve True si el string es num칠rico (solo d칤gitos)
print(animal.isdigit())  # False

# Devuelve True si el string est치 en min칰sculas
print(animal.islower())  # True


## Secuencias de Escape en Python

En Python, una secuencia de escape es una combinaci칩n de caracteres que no representa su valor literal, sino que tiene un significado especial para el int칠rprete de Python. Las secuencias de escape son 칰tiles cuando necesitas incluir caracteres especiales en cadenas de texto, como comillas, saltos de l칤nea, tabulaciones, y m치s.

A continuaci칩n, se describen algunas de las secuencias de escape m치s comunes con ejemplos pr치cticos.

### Ejemplos de Secuencias de Escape

1. **Comillas Dobles (`\"`)**: Permite incluir comillas dobles dentro de una cadena de texto delimitada por comillas dobles.

   ```python
   print("Ella dijo: \"Hola Mundo\"")
   # Salida: Ella dijo: "Hola Mundo"
   ```

2. **Comillas Simples (`\'`)**: Permite incluir comillas simples dentro de una cadena de texto delimitada por comillas simples.

   ```python
   print('It\'s a beautiful day!')
   # Salida: It's a beautiful day!
   ```

3. **Barra Invertida (`\\`)**: Permite incluir una barra invertida en la cadena.

   ```python
   print("Este es un signo de barra invertida: \\")
   # Salida: Este es un signo de barra invertida: \
   ```

4. **Salto de L칤nea (`\n`)**: Inserta un salto de l칤nea en la cadena.

   ```python
   print("Hola\nMundo")
   # Salida:
   # Hola
   # Mundo
   ```

5. **Tabulador (`\t`)**: Inserta un tabulador en la cadena.

   ```python
   print("Hola\tMundo")
   # Salida: Hola   Mundo
   ```

6. **Retroceso (`\b`)**: Elimina el car치cter anterior en la cadena.

   ```python
   print("Hola Mundoo\b")
   # Salida: Hola Mundo
   ```

7. **Retorno de Carro (`\r`)**: Mueve el cursor al principio de la l칤nea, sobrescribiendo el texto.

   ```python
   print("12345\rabc")
   # Salida: abc45
   ```

8. **Salto de P치gina (`\f`)**: Inserta un salto de p치gina en la cadena.

   ```python
   print("Primera p치gina\fSegunda p치gina")
   # Salida:
   # Primera p치gina
   #             Segunda p치gina
   ```

9. **Tabulador Vertical (`\v`)**: Inserta un tabulador vertical en la cadena.

   ```python
   print("Hola\vMundo")
   # Salida: Hola
   #         Mundo
   ```

10. **Beep (`\a`)**: Produce un sonido de beep (si el sistema lo soporta).

    ```python
    print("\a")
    ```

11. **Unicode (`\uXXXX`, `\UXXXXXXXX`, `\xXX`)**: Inserta caracteres Unicode en la cadena.

    ```python
    print("Hola \u263A")
    # Salida: Hola 驕죞n    ```

12. **Secuencia Octal (`\ooo`)**: Representa un car치cter utilizando una secuencia octal.

    ```python
    print("\101")
    # Salida: A
    ```

13. **Caracter Nulo (`\0`)**: Representa el car치cter nulo, aunque no tiene un efecto visible en la mayor칤a de las salidas.

    ```python
    print("Hola\0Mundo")
    # Salida: HolaMundo
    ```

14. **Nombre Unicode (`\N{name}`)**: Inserta un car치cter Unicode utilizando su nombre oficial.

    ```python
    colegio = "Villavicencio \N{smiling face with smiling eyes} \"La Salle\""
    print(colegio)
    # Salida: Villavicencio 游땕 "La Salle"
    ```

### Resumen

Las secuencias de escape son esenciales para manejar texto de manera m치s flexible y precisa en Python. Ya sea que necesites manejar comillas dentro de cadenas, controlar el formato del texto con saltos de l칤nea o tabulaciones, o insertar caracteres especiales como emojis, las secuencias de escape te ofrecen las herramientas necesarias.

## N칰meros en Python

Python es un lenguaje vers치til que maneja varios tipos de n칰meros, desde enteros y decimales hasta n칰meros complejos. A continuaci칩n, exploraremos los diferentes tipos de n칰meros en Python, sus propiedades, conversiones y las operaciones b치sicas que se pueden realizar con ellos.

### Tipos de N칰meros en Python

1. **Enteros (`int`)**: Son n칰meros sin parte decimal. Ejemplo:
   ```python
   entero = 2
   ```

2. **Decimales (`float`)**: Son n칰meros que incluyen una parte decimal. Ejemplo:
   ```python
   decimal = 2.5
   ```

3. **Complejos (`complex`)**: Son n칰meros que tienen una parte real y una parte imaginaria. Ejemplo:
   ```python
   complejo = 2 + 5j
   ```

### Propiedades de los N칰meros

Puedes utilizar la funci칩n `type()` para verificar el tipo de dato de un n칰mero:

```python
print(type(entero))   # <class 'int'>
print(type(decimal))  # <class 'float'>
print(type(complejo)) # <class 'complex'>
```

### Conversi칩n de Tipos

Python permite convertir un tipo de n칰mero a otro. Aqu칤 algunos ejemplos de conversiones comunes:

```python
print(float(entero))  # Convierte el entero a decimal: 2.0
print(int(decimal))   # Convierte el decimal a entero: 2
print(complejo.real)  # Obtiene la parte real del n칰mero complejo: 2.0
print(complejo.imag)  # Obtiene la parte imaginaria del n칰mero complejo: 5.0
```

### Operaciones con Asignaci칩n

Python permite realizar operaciones y asignar el resultado a una variable en un solo paso. Aqu칤 algunos ejemplos:
