# **Rutas Útiles y Rápidas en Visual Studio Code**

## **Selección y Edición de Código**

- **Ctrl + }**: Comentar y descomentar líneas de código.
  - Este atajo es muy útil cuando necesitas agregar o quitar comentarios en múltiples líneas de código de manera rápida.

- **Ctrl + Shift + L**: Seleccionar todas las ocurrencias de una palabra o variable en el archivo.
  - Ideal para realizar cambios en todas las instancias de una palabra o variable a la vez.

- **Ctrl + D**: Seleccionar la siguiente ocurrencia de la palabra o variable seleccionada.
  - Este atajo te permite seleccionar y modificar múltiples ocurrencias una a la vez, de forma progresiva.

## **Navegación Rápida**

- **Ctrl + P**: Abrir rápidamente un archivo por su nombre.
  - Simplemente comienza a escribir el nombre del archivo y VS Code te mostrará las coincidencias.

- **Ctrl + G**: Ir a una línea específica.
  - Muy útil cuando recibes un mensaje de error que indica en qué línea del código se encuentra un problema.

- **Ctrl + Shift + O**: Navegar entre los símbolos del archivo (como funciones o clases).
  - Permite moverse rápidamente entre diferentes partes del código.

## **Manipulación de Código**

- **Alt + Shift + Flecha Arriba/Flecha Abajo**: Duplicar la línea actual o las líneas seleccionadas hacia arriba o hacia abajo.
  - Ideal para copiar rápidamente una línea o bloque de código sin tener que usar copiar y pegar.

- **Ctrl + Shift + K**: Eliminar la línea actual.
  - Este atajo elimina una línea de código sin necesidad de seleccionarla primero.

- **Ctrl + /**: Comentar o descomentar la línea actual o la selección.
  - Este es otro método rápido para comentar o descomentar líneas de código.

## **Búsqueda y Reemplazo**

- **Ctrl + F**: Abrir el cuadro de búsqueda.
  - Permite buscar palabras o patrones en el archivo actual.

- **Ctrl + H**: Abrir el cuadro de búsqueda y reemplazo.
  - Facilita la búsqueda y reemplazo de texto en todo el archivo o en una selección.

- **F3 / Shift + F3**: Buscar la siguiente o anterior ocurrencia.
  - Navega fácilmente a través de las ocurrencias de una búsqueda en el archivo.

## **Gestión de Archivos**

- **Ctrl + N**: Crear un nuevo archivo.
  - Rápido acceso para comenzar un nuevo archivo desde cero.

- **Ctrl + S**: Guardar el archivo actual.
  - Asegúrate de que los cambios se guarden con este atajo común.

- **Ctrl + W**: Cerrar la pestaña actual.
  - Cierra rápidamente la pestaña actual sin necesidad de usar el ratón.

## **Visualización**

- **Ctrl + B**: Mostrar u ocultar la barra lateral.
  - Maximiza el espacio de trabajo ocultando la barra lateral cuando no la necesites.

- **Ctrl + `**: Abrir o cerrar la terminal integrada.
  - Accede a la terminal sin salir del editor de código.

- **Ctrl + Shift + E**: Mostrar el explorador de archivos.
  - Vuelve rápidamente al explorador de archivos para gestionar tus proyectos.

## **Depuración**

- **F5**: Iniciar la depuración.
  - Ejecuta tu código en modo de depuración.

- **F9**: Alternar un breakpoint en la línea actual.
  - Coloca o quita un punto de interrupción para depurar.

- **F10**: Paso por paso (Step Over) en la depuración.
  - Avanza la ejecución línea por línea, sin entrar en funciones.

## **Productividad**

- **Ctrl + Shift + P**: Abrir la paleta de comandos.
  - Accede a todas las funcionalidades de VS Code desde un solo lugar, solo tienes que empezar a escribir.

- **Ctrl + Tab**: Cambiar entre archivos abiertos.
  - Navega rápidamente entre los archivos abiertos en tus pestañas.

- **Alt + Click**: Crear múltiples cursores.
  - Añade múltiples cursores para editar texto en varios lugares a la vez.

## **Conclusión**

Estos atajos y rutas rápidas te ayudarán a mejorar tu eficiencia y flujo de trabajo en Visual Studio Code. Memorizar y practicar estos comandos te permitirá realizar tareas comunes con mayor rapidez, lo que te permitirá concentrarte más en escribir y depurar tu código.
